```python
def flux_to_lum(flux, fluxerr, dist, e_dist):
    lum = 1.0e24*1.2*flux*dist*dist
    lumerr = 1.0e24*np.sqrt((2.4*flux*dist*e_dist)**2.0 + (1.2*dist*dist*fluxerr)**2.0)
    return (lum, lumerr)

g19_lum, g19_lumerr = flux_to_lum(data['flux']*1.0e3, data['flux_err']*1.0e3, 8.5e-3, 0.0)

density = np.array([0.04]) #cm^{-3}, what is the density estimate?
freq = 0.84e9 #GHz
```

### Statistical test 

```python
import snrlightcurve_freqparam as snr_ind
reload(snr_ind)

def lum_exp(n0, mej, ekin, ages):
    lum = np.zeros(ages.size)
    for i, age_i in enumerate(ages):
        rad, lum[i], vel = snr_ind.lightcurve(n0=n0, mej=mej, e51=ekin, epse=0.0001, pp=2.2,\
                                                       freq=freq, tsnap=age_i, sntype='ia')
    return lum

ages = data['age']-1868

start = tm.time()
e51array = np.logspace(-1, 1, 500)
mejarray = np.linspace(0.1,2.0,500)
chisq_lum = np.zeros((e51array.size, mejarray.size))

for i, e51 in enumerate(e51array):
    for j, mej in enumerate(mejarray):
        chisq_lum[i,j], pval = chisquare(g19_lum/1.0e22, lum_exp(density[0], mej, e51, ages)/1.0e22)


       
print 'RUNTIME = ',str(dt.timedelta(seconds=tm.time()-start))
```

```python
M, e = np.meshgrid(mejarray, e51array)

#~~~~~~INPUT~~~~~~~~#
mej = 1.4
e51 = 2.5
n0=0.04
showInputOnContourPlot = True
#~~~~~~~~~~~~~~~~~~~#

plt.figure(figsize=(7,6))
cs = plt.contourf(M,e,np.log10(chisq_lum),9, cmap='gray')
plt.colorbar()
plt.plot([1.4], [1.0], 'rx', mew=2.0, ms=10.0)
plt.text(1.2, 1.16, 'Normal TypeIa', fontsize=15)

if showInputOnContourPlot:
    plt.axvline(x=mej, ls='-', color='r', lw=1.0)
    plt.axhline(y=e51, ls='-', color='r', lw=1.0)
    plt.plot([mej], [e51], 'ro')
plt.yscale('log')
plt.ylim(0.1,5.0)
plt.xlabel(r'Ejecta Mass [$\rm{M_{\odot}}$]', fontsize=20)
plt.ylabel(r'Kinetic Energy [$\rm{\times\ 10^{51}\ ergs}$]', fontsize=20)
plt.show()


t, r_t, lum_t, v_t = snr_ind.lightcurve_Full(n0=n0, e51=e51, mej=mej, pp=2.2, \
                                         freq=freq, epse=1.0e-4, sntype='ia')


plt.rc('font', family='serif')
plt.figure(figsize=(6,5))

plt.plot(t[np.nonzero(lum_t)], lum_t[np.nonzero(lum_t)]/1.0e22, 'k-')

plt.errorbar(data['age']-1868., g19_lum/1.0e22, yerr=g19_lumerr/1.0e22, fmt='.')

plt.xscale('log')
#plt.yscale('log')
plt.xlabel('Age [years]', fontsize=18)
plt.ylabel('$L_{0.84}$ [$\\times$ $10^{22}$ ergs/s/Hz]', fontsize=18)
#plt.legend(loc=2, frameon=False, fontsize=13)
plt.tight_layout()
plt.xlim(100, 150)
plt.ylim(0,20)
#plt.savefig('density_mej_example.pdf')
plt.show()
```
